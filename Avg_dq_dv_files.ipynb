```python
import glob
import os
import numpy as np
from collections import defaultdict
def get_files_list(start_id, end_id, file_paths):   
    temp = defaultdict(list)
    for i in range(start_id, end_id +1):
        temp[str(i)] = [f for f in file_paths if os.path.basename(f).split("_")[0] == str(i)]
        
    return temp 
```

```python
file_paths = [file for file in glob.glob(os.path.join(folder_path, '**/*.xls'), recursive=True)]
files_path_dict = get_files_list(1,39, file_paths)
print(files_path_dict)
```

```python
file_paths_dict = {}
for key, value in files_path_dict.items():
    sorted_paths = sorted(value, key=lambda x: int(os.path.basename(x).split("_")[-1].split(".")[0]))
    file_paths_dict[key] = sorted_paths
```

```python
import pandas as pd
for cell_id,paths in file_paths_dict.items():
    #if int(cell_id) == 2:
        #break
    appended_df = pd.DataFrame()
    for file_path in paths:
        # read all the sheets in the Excel file
        xl = pd.ExcelFile(file_path)
        # check if both sheets start with "Detail"
        sheet_names = [name for name in xl.sheet_names if name.startswith("Detail")]
        if len(sheet_names) == 2:
            # merge the two sheets and save to a new Excel file with the same name
            df1 = pd.read_excel(xl, sheet_names[0])
            df2 = pd.read_excel(xl, sheet_names[1])
            merged_df = pd.concat([df1, df2], ignore_index=True)
            max_value = merged_df['Cycle'].max()
            merged_df['dq/dv'] = (merged_df['CapaCity(mAh)'] - merged_df['CapaCity(mAh)'].shift(1))/(merged_df['Voltage(V)'] - merged_df['Voltage(V)'].shift(1))
            #merged_df.to_excel(merged_file_path, index=False)
            
            for i in range(max_value):
                merged_df_data = merged_df[(merged_df["Status"] == "CC_DChg") & (merged_df["Cycle"] == i)]
                appended_df = pd.concat([appended_df, merged_df_data])
        elif len(sheet_names) < 2:
            # add the details of the skipped file to the DataFrame
            skipped_files_df = pd.read_excel(xl, sheet_names[0])
            max_value_2 = skipped_files_df['Cycle'].max()
            skipped_files_df['dq/dv'] = (skipped_files_df['CapaCity(mAh)'] - skipped_files_df['CapaCity(mAh)'].shift(1))/(skipped_files_df['Voltage(V)'] - skipped_files_df['Voltage(V)'].shift(1))
            
            for i in range(max_value_2):
                skipped_df_data = skipped_files_df[(skipped_files_df["Status"] == "CC_DChg") & (skipped_files_df["Cycle"] == i)]
                appended_df = pd.concat([appended_df, skipped_df_data])
                
    # for current cell
    file_save_path = os.path.join(cell_merged_files_directory, f"{cell_id}_merged.xlsx")
    # convert column to numeric type
    appended_df['dq/dv'] = pd.to_numeric(appended_df['dq/dv'], errors='coerce')
    
    # replace 'inf' values with NaN
    appended_df['dq/dv'] = appended_df['dq/dv'].replace([np.inf, -np.inf], np.nan)
    
    # fill NaN values with previous row's value
    appended_df['dq/dv'] = appended_df['dq/dv'].fillna(method='ffill')
    appended_df.to_excel(file_save_path, index=False)
```

```python
import os
from openpyxl import load_workbook

# Iterate over the files in the folder
for filename in os.listdir(cell_merged_files_directory):
    if filename.endswith(".xlsx"):  # Assuming all files are in .xlsx format
        file_path = os.path.join(cell_merged_files_directory, filename)
        workbook = load_workbook(file_path)
        
        # Select the active sheet
        sheet = workbook.active
        # Get the number of rows
        num_rows = sheet.max_row
        sheet.cell(row=1, column=14).value = 1
        sheet.cell(row=1, column=13).value = "Avg. dq/dv"
        sheet.cell(row=1, column=14).value = "Cycles"
        start_row = 2
        cycle = 0
        end_row = 26
        answer_row = 26
        
        # Iterate through the rows and calculate the averages
        for row in range(2, num_rows + 1):
            dq_dv = sheet.cell(row=row, column=12).value
            # Calculate the average for the previous cycle
            average_formula = f"=AVERAGE(L{start_row}:L{end_row})"
            sheet.cell(row=answer_row, column=13).value = average_formula
            end_row = end_row + 1
            start_row = start_row + 1
            answer_row = answer_row + 1

        for row in range(2, num_rows + 1):
            dq_dv = sheet.cell(row=row, column=12).value
            if dq_dv == 0:
                for col in range(row, row + 24):
                    cell = sheet.cell(row=col, column=13)
                    cell.value = None

        for row in range(2, num_rows + 1):
            dq_dv = sheet.cell(row=row, column=12).value
            if dq_dv != 0:
                cycles_number = cycle
                sheet.cell(row=row, column=14).value = cycles_number
            else:
                cycle = cycle + 1
                sheet.cell(row=row, column=14).value = cycle

        # Save the workbook with the original name appended with "_output.xlsx"
        output_filename = filename[:-5] + "_output.xlsx"  # Add "_output" before the extension
        output_file_path = os.path.join(avg_dq_dv_files_directory, output_filename)
        workbook.save(output_file_path)
```
